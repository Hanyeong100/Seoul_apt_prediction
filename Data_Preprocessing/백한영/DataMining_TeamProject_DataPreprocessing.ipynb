# 서울특별시 부동산 실거래가 데이터 준비

```python
sales = pd.read_csv('/content/서울시 부동산 실거래가 정보.csv', encoding='cp949')
```

```python
sales_sorted = sales.sort_values('계약일')
```

```python
sales_sorted.columns
```

```python
# 각종 거래 관련 결측치 제외 (거래가격, 면적, 건축년도 등)
print(sales_sorted.shape)
print(sales_sorted.info())
```

```python
# 거래 취소 매물 제외
sales_sorted = sales_sorted[sales_sorted['취소일'].isnull()]
sales_sorted = sales_sorted[sales_sorted['물건금액(만원)'].notnull()]
sales_sorted = sales_sorted[sales_sorted['건축년도'].notnull()]
sales_sorted = sales_sorted[sales_sorted['건물명'].notnull()]
sales_sorted = sales_sorted[sales_sorted['건물용도']=='아파트']
sales_sorted = sales_sorted[sales_sorted['건축년도']>1900]
```

```python
sales_sorted = sales_sorted[['계약일', '자치구코드', '자치구명', '법정동코드', '법정동명', '지번구분', '지번구분명', '본번', '부번', '건물명', '건물면적(㎡)', '토지면적(㎡)', '물건금액(만원)', '건축년도', '건물용도']]
```

```python
sales_sorted
```

```python
sales_sorted.to_csv('부동산 실거래가(2016~2023).csv', encoding = 'utf-8-sig')
```
