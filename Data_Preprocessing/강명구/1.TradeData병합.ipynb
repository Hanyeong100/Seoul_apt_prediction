```python
import os
```

```python
listdir = os.listdir('./Data/Trade/')
path = './Data/Trade/'
listdir
```

```python
import pandas as pd
```

```python
df = pd.read_csv(path+listdir[0], encoding='euc-kr')
df.head()
```

```python
apt_trade = pd.DataFrame()

for csv in listdir:
    df = pd.read_csv(path+csv, encoding='euc-kr') # 데이터 하나씩 읽기
    df = df[df['건물용도'] == '아파트'] # 아파트만 추출
    apt_trade = pd.concat([apt_trade, df], axis=0)
```

```python
apt_trade.to_csv('./Data/apt_trade.csv', encoding='euc-kr')
```
