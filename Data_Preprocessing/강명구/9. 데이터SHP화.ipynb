```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon, LineString
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.font_manager as fm
```

```python
IC = pd.read_csv('./Data/1차정제/IC위치데이터.csv', encoding='euc-kr')
```

```python
IC
```

```python
IC['경도'] = IC['경도'].astype(float)
IC['위도'] = IC['위도'].astype(float)
IC['geometry'] = IC.apply(lambda row : Point([row['경도'], row['위도']]), axis=1)
IC = gpd.GeoDataFrame(IC, geometry='geometry')
IC.crs = {'init':'epsg:4326'}
IC.to_file('./Data/1차정제/GEO/서울시IC위치.shp', encoding='euc-kr')
```
