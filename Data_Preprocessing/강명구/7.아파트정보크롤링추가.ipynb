```python
def remove_text_inside_parentheses(text_list):
    result = []
    for text in text_list:
        while "(" in text and ")" in text:
            start_index = text.index("(")
            end_index = text.index(")")
            text = text[:start_index] + text[end_index+1:]
        result.append(text)
    return result
```
